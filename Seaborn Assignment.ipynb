```python
'''Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Answer:- The Seaborn library in Python is a powerful and versatile visualization library built on top of Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. Here are five plots that can be created using Seaborn, along with their uses:

1. Scatter Plot: A scatter plot is used to display the relationship between two numerical variables. Each point represents an observation in the dataset.
2. Line Plot: A line plot is used to display data points along a continuous variable, often used to track changes over time.
3. Box Plot: A box plot (or whisker plot) is used to show the distribution of a dataset based on a five-number summary: minimum, first quartile, median, third quartile, and maximum. It is useful for detecting outliers and comparing distributions.
4. Histogram: A histogram is used to display the distribution of a single numerical variable by dividing the data into bins and counting the number of observations in each bin.
5. Heatmap: A heatmap is used to display the relationship between two categorical variables. It is often used to show correlations between variables in a matrix format where the intensity of color represents the value.
'''
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

tips = sns.load_dataset("tips")
flights = sns.load_dataset("flights")

plt.figure(figsize=(10, 6))
plt.subplot(3, 2, 1)
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.title('Scatter Plot of Total Bill vs Tip')

plt.subplot(3, 2, 2)
sns.lineplot(data=tips, x="total_bill", y="tip")
plt.title('Line Plot of Total Bill vs Tip')

plt.subplot(3, 2, 3)
sns.boxplot(data=tips, x="day", y="total_bill")
plt.title('Box Plot of Total Bill by Day')

plt.subplot(3, 2, 4)
sns.histplot(data=tips, x="total_bill", bins=20, kde=True)
plt.title('Histogram of Total Bill')

flights_pivot = flights.pivot(index="month", columns="year", values="passengers")

plt.subplot(3, 2, 5)
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Heatmap of Flights Passengers')

plt.tight_layout()
plt.show()
```

```python
#Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.

fmri = sns.load_dataset("fmri")

plt.figure(figsize=(12, 6))
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
plt.title('FMRI Signal over Time')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.legend(title='Event and Region')
plt.show()
```

```python
'''Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.

Note: pclass, age, and fare are columns in the titanic dataset.

'''
titanic = sns.load_dataset("titanic")

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sns.boxplot(ax=axes[0], data=titanic, x='pclass', y='age')
axes[0].set_title('Age vs Passenger Class')
axes[0].set_xlabel('Passenger Class')
axes[0].set_ylabel('Age')

sns.boxplot(ax=axes[1], data=titanic, x='pclass', y='fare')
axes[1].set_title('Fare vs Passenger Class')
axes[1].set_xlabel('Passenger Class')
axes[1].set_ylabel('Fare')

plt.tight_layout()
plt.show()
```

```python
'''Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.'''

diamonds = sns.load_dataset("diamonds")

plt.figure(figsize=(10,6))
sns.histplot(data=diamonds, x="price", hue="cut", multiple="stack", bins=30)
plt.title('Histogram of  Diamond Prices with Cut as Hue')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

```python
'''Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.'''

iris = sns.load_dataset("iris")

plt.figure(figsize=(10,6))
sns.pairplot(data=iris, hue="species")
plt.suptitle('Pair Plot of Iris Dataset',y=1.02)
plt.show()
```

```python
#Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

flights = sns.load_dataset("flights")

flights_pivot = flights.pivot(index="month", columns="year", values="passengers")

plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Heatmap of Flights Dataset')
plt.show()
```
