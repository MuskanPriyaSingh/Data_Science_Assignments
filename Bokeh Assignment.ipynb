```python
'''Q1. How can you create a Bokeh plot using Python code?

Answer:- Creating a Bokeh plot using Python involves a few key steps:

1. Install Bokeh if you haven't already.
2. Import the necessary functions and modules from Bokeh.
3. Prepare your data.
4. Create a figure object.
5. Add glyphs (such as lines, circles, bars) to the figure.
6. Customize the plot as needed.
7. Show or save the plot.
'''

from bokeh.plotting import figure, show, output_notebook
from bokeh.io import output_file

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

output_notebook()

output_file('line_plot.html')
p = figure(title = 'Line Plot', x_axis_label = 'x', y_axis_label ='y')
p.line(x, y, legend_label = 'Temp.', line_width = 2)
show(p)
```

```python
'''Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example

Answer:- In Bokeh, glyphs are the visual shapes that represent data on a plot, such as circles, lines, rectangles, or any other shapes. They are the fundamental building blocks used to create visual representations of data. You can add glyphs to a Bokeh plot using various glyph methods provided by Bokeh, such as circle, line, rect, triangle, etc.
'''
output_notebook()

p = figure(title="Example of Glyphs in Bokeh", x_axis_label='x', y_axis_label='y')

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Add circle glyphs
p.circle(x, y, size=10, color="navy", alpha=0.5, legend_label="Circle Glyphs")

# Add line glyphs
p.line(x, y, legend_label="Line Glyphs", line_width=2)

# Add triangle glyphs
p.triangle(x, [i+1 for i in y], size=10, color="firebrick", alpha=0.5, legend_label="Triangle Glyphs")

# Add square glyphs
p.square(x, [i-1 for i in y], size=10, color="green", alpha=0.5, legend_label="Square Glyphs")

show(p)
```

```python
'''Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

Answer:- To customize the appearance of a Bokeh plot, including the axes, title, and legend, you can use various attributes and methods provided by Bokeh. Here's an example that demonstrates how to customize these elements:

1. Title: You can set the title of the plot using the title attribute.
2. Axes: Customize axes by modifying properties like axis_label, axis_line_width, major_label_text_font_size, major_tick_line_color, etc.
3. Legend: Customize the legend using properties like label_text_font_size, label_text_font_style, background_fill_color, etc.
'''
output_notebook()

# Create a new plot with a title
p = figure(title="Customized Bokeh Plot", width=700, height=500)

# Add a line renderer with legend and line thickness
line1 = p.line([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], line_width=2, color='blue', legend_label="Line 1")
line2 = p.line([1, 2, 3, 4, 5], [4, 6, 3, 8, 7], line_width=2, color='red', legend_label="Line 2")

# Customize the title
p.title.text = "Customized Bokeh Plot"
p.title.align = "center"
p.title.text_font_size = "20pt"
p.title.text_color = "green"
p.title.text_font_style = "italic"

# Customize the x-axis
p.xaxis.axis_label = "X-Axis Label"
p.xaxis.axis_label_text_color = "blue"
p.xaxis.axis_label_text_font_size = "14pt"
p.xaxis.major_label_text_font_size = "12pt"
p.xaxis.major_label_text_color = "purple"
p.xaxis.axis_line_width = 2

# Customize the y-axis
p.yaxis.axis_label = "Y-Axis Label"
p.yaxis.axis_label_text_color = "red"
p.yaxis.axis_label_text_font_size = "14pt"
p.yaxis.major_label_text_font_size = "12pt"
p.yaxis.major_label_text_color = "orange"
p.yaxis.axis_line_width = 2

# Customize the legend
p.legend.label_text_font_size = "12pt"
p.legend.label_text_font_style = "bold"
p.legend.background_fill_color = "lightgray"
p.legend.border_line_width = 2
p.legend.border_line_color = "black"
p.legend.location = "top_left"

show(p)
```
