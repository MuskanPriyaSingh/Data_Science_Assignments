```python
'''Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Answer:- Matplotlib is a popular plotting library in Python used for creating static, interactive, and animated visualizations. It is widely used for data visualization due to its versatility and the ability to produce high-quality plots.

Five plots that can be plotted using the Pyplot module of Matplotlib:

1. Line Plot: Used to display data points connected by straight lines. It is useful for visualizing trends over time or continuous data.
2. Bar Plot: Used to represent data with rectangular bars. The length of each bar is proportional to the value it represents.
3. Histogram: Used to represent the distribution of numerical data by dividing the data into bins and plotting the frequency of each bin.
4. Scatter Plot: Used to display values for typically two variables for a set of data. Points are plotted on a Cartesian coordinate system.
5. Pie Chart: Used to represent data in a circular graph, where each slice of the pie represents a proportion of the whole.
'''
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 12))

plt.subplot(3, 2, 1)
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.title("Line Plot")

plt.subplot(3, 2, 2)
plt.bar([1, 2, 3, 4], [10, 20, 15, 25])
plt.title("Bar Plot")

plt.subplot(3, 2, 3)
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
plt.hist(data, bins=4)
plt.title("Histogram")

plt.subplot(3, 2, 4)
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.scatter(x, y)
plt.title("Scatter Plot")

plt.subplot(3, 2, 5)
sizes = [10, 20, 30, 40]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart")

plt.tight_layout

plt.show()
```

```python
'''Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

Answer:- A scatter plot is a type of data visualization that displays values for typically two variables for a set of data. Each point on the plot represents an observation in the dataset, with the position on the x-axis corresponding to the value of one variable and the position on the y-axis corresponding to the value of the other variable. Scatter plots are useful for identifying relationships or correlations between the two variables.
'''

import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50) 
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x, y)

plt.title("Scatter Plot of Generated Data")
plt.xlabel("x-values")
plt.ylabel("y-values")

plt.show()

#Note: Also add title, xlabel, and ylabel to the plot.
```

```python
'''Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

Answer:- The 'subplot()' function in Matplotlib is used to create multiple plots within a single figure. This function allows you to arrange multiple plots in a grid layout, making it easier to compare different datasets or visualize multiple aspects of the same data in one view.
'''
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.figure(figsize=(10, 8))

# Plot 1
plt.subplot(2, 2, 1)  # specifies a 2x2 grid and selects the first subplot
plt.plot(x1, y1)
plt.title('Line 1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Plot 2
plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title('Line 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Plot 3
plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title('Line 3')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Plot 4
plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title('Line 4')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.tight_layout()

plt.show()
```

```python
'''Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

Answer:- A bar plot is a chart that presents categorical data with rectangular bars with lengths proportional to the values that they represent. Bar plots are used to compare different categories or to track changes over time when the changes are larger.

It's used for:
-> To compare different groups or categories.
-> To show changes or trends over time for categorical data.
-> To make it easier to visualize and understand the distribution and relationships between different variables.
'''
company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.bar(company, profit, color='blue')
plt.title('Company Profit - Vertical Bar Plot')
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')

plt.subplot(1, 2, 2)
plt.barh(company, profit, color='green')
plt.title('Company Profit - Horizontal Bar Plot')
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')

plt.tight_layout()

plt.show()
```

```python
'''Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

Answer:- A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays the dataset's minimum, first quartile (Q1), median, third quartile (Q3), and maximum values. Here's a breakdown of what each part represents:

1. Box: Represents the interquartile range (IQR) from Q1 to Q3, which contains the middle 50% of the data.
2. Whiskers: Extend from the ends of the box to the minimum and maximum values within a certain range (often 1.5 times the IQR).
3. Median Line: A line within the box that shows the median value (the middle of the data).
4. Outliers: Data points outside the whiskers are considered outliers.

Box plots are useful for:
-> Comparing distributions between multiple groups.
-> Identifying outliers.
-> Understanding the spread and skewness of the data.
'''

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot([box1, box2], labels=['box1', 'box2'])

plt.xlabel('Groups')
plt.ylabel('Values')
plt.title('Box Plot of Two Datasets')

plt.show()
```
